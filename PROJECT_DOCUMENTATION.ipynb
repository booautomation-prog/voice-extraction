## 🔧 Core Components Explained

### 1. **app.py - Flask Web Server**
- **Lines:** ~180
- **Purpose:** Main backend server handling all requests
- **Key Routes:**
  - `GET /` - Serve homepage
  - `POST /api/download` - Start download job
  - `GET /api/status/<job_id>` - Check job progress
  - `GET /api/stream-stem/<job_id>/<stem_name>` - Stream audio
  - `GET /api/download-stem/<job_id>/<stem_name>` - Download stem
- **Features:**
  - Background thread processing
  - Job queue management
  - Error handling & logging

### 2. **download_audio.py - YouTube Downloader**
- **Lines:** ~120
- **Purpose:** Download audio from YouTube URLs
- **Dependencies:** yt-dlp, FFmpeg
- **Output:** MP3 files (192 kbps)
- **Batch Support:** Can process multiple URLs from file

### 3. **separate_audio.py - Audio Separator**
- **Lines:** ~140
- **Purpose:** Separate audio into 4 stems
- **AI Model:** Demucs (by Meta)
- **Output Stems:**
  - `vocals.wav` - Lead vocals
  - `drums.wav` - Drum tracks
  - `bass.wav` - Bass lines
  - `other.wav` - Other instruments
- **Model Options:**
  - `mdx` (default) - Faster, cloud-friendly
  - `htdemucs` - Higher quality
  - `mdx_extra` - Highest quality, slowest

## 🛠️ Local Development Setup

### **Prerequisites**
1. Python 3.11+
2. FFmpeg installed
3. Git

### **Installation**
```bash
# Clone repository
git clone https://github.com/booautomation-prog/voice-extraction.git
cd voice-extraction

# Create virtual environment (optional but recommended)
python -m venv venv
source venv/bin/activate  # or: venv\Scripts\activate on Windows

# Install dependencies
pip install -r requirements.txt
```

### **Running Locally**
```bash
# Development server
python app.py

# Access at: http://localhost:5000
```

### **Command Line Usage**
```bash
# Download only
python download_audio.py "https://youtube.com/watch?v=..." -o my_music

# Separate audio
python separate_audio.py "song.mp3" -m htdemucs

# Batch process
python download_audio.py -f urls.txt
```

## ⚠️ Known Limitations & Solutions

| Issue | Cause | Solution |
|-------|-------|----------|
| Long processing time | AI model inference | Shorter songs (~5 min) process faster |
| High RAM usage | PyTorch + Demucs | Normal for GPU ML, expected on Railway |
| Model download timeout | Large model (~300MB) | Pre-cached in Docker |
| Slow on CPU-only | GPU required for speed | Railway uses CPU on free tier |
| Storage limits | Free tier constraints | Auto-cleanup of old jobs recommended |
| YouTube age-restricted | yt-dlp limitation | Use different videos |
| Audio quality | Separation algorithm | Use higher quality input for better output |
